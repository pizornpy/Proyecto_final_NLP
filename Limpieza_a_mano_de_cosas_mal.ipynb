```python
df = pd.read_csv(r"C:\Users\Juan\Desktop\Proyecto_final_NLP\df_final.csv")
```

```python
df['school'] = df['school'].str.replace('[', '').str.replace(']', '').str.replace("'", '')
```

```python
df.loc[df['author'] == 'Hegel', 'school'] = 'german_idealism'
df.loc[df['author'] == 'Hegel', 'school'] = 'german_idealism'
```

```python
grouped_df = df.groupby('title').agg({
    'sentence_lowered': ' '.join,
    'author': "first",
    'school': 'first',
    'dif': 'mean'
}).reset_index()
```
