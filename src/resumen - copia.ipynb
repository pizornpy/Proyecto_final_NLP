```python
vectorizer = TfidfVectorizer()
```

```python
def resumen(str):
    texto = str

    oraciones = sent_tokenize(texto)# Tokenizar el texto en oraciones

    # Lista de palabras adicionales para eliminar
    extra_stop_words = [",", ".", "--", ";", "``", ")", "(", "”", "“", "?", "‘", "’", "!", "{"]

    # Tokenizar el texto en palabras y eliminar palabras irrelevantes (stop words)
    stop_words = set(stopwords.words('english') + extra_stop_words)
    word_tokens = [word_tokenize(sentence) for sentence in oraciones]
    palabras_filtradas = [[word.lower() for word in words if word.lower() not in stop_words] for words in word_tokens]

    # Convertir las palabras nuevamente a oraciones
    filtered_oraciones = [' '.join(words) for words in palabras_filtradas]


    tfidf_matrix = vectorizer.fit_transform(filtered_oraciones) # Obtener la matriz TF-IDF

    scores = []
    for i in range(len(filtered_oraciones)):    # Calcular la importancia de las palabras en cada oración y seleccionar las más relevantes

        sentence = filtered_oraciones[i]
        doc = tfidf_matrix[i]  
        feature_index = doc.nonzero()[1]
        tfidf_scores = zip(feature_index, [doc[0, x] for x in feature_index])
        score_sum = sum([score for _, score in tfidf_scores])
        scores.append((sentence, score_sum))

    # Ordenar las oraciones según su importancia en el TF-IDF
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    N = 2
    res_orac = [o for o, _ in scores[:N]] # Seleccionar las primeras N oraciones para el resumen

    res = ' '.join(res_orac)
    return res

```

```python
df["resumen"] = df["work"].map(resumen)
```

```python
df.to_csv(r"C:\Users\Juan\Desktop\Proyecto_final_NLP\datares.csv", index=False)
```
