```python
X = og["sentence_lowered"]
y = og["school"]


X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.2, random_state = 42)
```

```python
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
num_classes = len(label_encoder.classes_)
```

```python
max_sequence_length = 300  # Maximum sequence length for padding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(res["resumen"])

# Create a new empty column for the transformed sequences
res["resumen_sequences"] = ""

# Iterate over each row
for index, row in res.iterrows():
    text = row["resumen"]
    
    # Tokenize the text
    sequences = tokenizer.texts_to_sequences([text])
    
    # Pad the sequences
    padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length)
    
    # Update the corresponding row with the transformed sequences
    res.at[index, "resumen_sequences"] = padded_sequences

# Print the updated DataFrame
```

```python
# Assuming you have a DataFrame called res with the transformed sequences in the "resumen_sequences" column
# and your model is named "model"

# Create a new empty column for the predictions
res["predictions"] = ""

# Iterate over each row
for index, row in res.iterrows():
    sequence = row["resumen_sequences"]
    
    # Reshape the sequence to match the expected input shape of the model
    sequence = sequence
    
    # Make a prediction using the model
    prediction = jacinto2.predict(sequence)
    
    # Update the corresponding row with the prediction
    res.at[index, "predictions"] = prediction

# Print the updated DataFrame
print(res)

```
